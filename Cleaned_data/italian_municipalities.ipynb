```python
import pandas as pd
import re
```

```python
# Reading Italian municipalities file and showing top 10
df_it_mun = pd.read_excel("../Datasets/Italian_municipalities.xlsx", engine="openpyxl")
df_it_mun.head(10)
```

```python
# Drop first 6 rows and reset indexing
df_it_mun = df_it_mun.iloc[6:]
df_it_mun.reset_index(drop=True, inplace=True)
df_it_mun.head(10)
```

```python
df_it_mun.columns = df_it_mun.iloc[0]  # Set the first row as column names
df_it_mun = df_it_mun[1:].reset_index(drop=True)  # Drop the first row and reset index
df_it_mun.head(10)
```

```python
df_it_mun = df_it_mun.iloc[:,:5] # Leave only first 5 columns
df_it_mun.head(10)
```

```python
df_it_mun = df_it_mun.drop(df_it_mun.columns[1], axis=1) # Drop NaN column
df_it_mun.head(10)
```

```python
# Rename columns
df_it_mun.columns.values[0] = "MunicipalityName"
df_it_mun.columns.values[1] = "RegionName"
df_it_mun.columns.values[2] = "AutoCode"
df_it_mun.columns.values[3] = "Pop2011"
df_it_mun.head(10)
```

```python
df_it_mun.isna().sum()/df_it_mun.shape[1] # Proportion of missing values for each column
```

```python
# Drop AutoCode column and Remove leading/trailing spaces
df_it_mun.columns = df_it_mun.columns.str.strip()
df_it_mun.drop("AutoCode", axis=1, inplace=True)
df_it_mun.head(10)
```

```python
df_it_mun.isna().sum()/df_it_mun.shape[1] # Proportion of missing values for each column
```

```python
df_it_mun.info()
```

```python
df_it_mun.dropna(inplace=True) # Drop missing values
df_it_mun.info()
```

```python
df_it_mun.tail(10) # Call for last 10 rows
```

```python
df_it_mun["RegionName"].unique() # Check for the unique values
```

```python
df_it_mun["RegionName"] = df_it_mun["RegionName"].str.strip().str.title() # Remove leading/trailing spaces and capitalize
```

```python
corrections = {
    "Piemonte": "Piemonte",
    "Piemonté": "Piemonte",
    "Looooombardia": "Lombardia",
    "Venetoooo": "Veneto",
    "Trentino-Alto Adige/Südtirollllll": "Trentino-Alto Adige/Südtirol",
    "Valle D'Aosta/Vallée D'Aosteeeeeee": "Valle D'Aosta/Vallée D'Aoste",
    "Emilia-Romagnaaaaa": "Emilia-Romagna",
    "Calabrià": "Calabria",
    "Siciliaaaaaa": "Sicilia",
    "Sardegnà": "Sardegna",
    "LIGURIA": "Liguria"
}

df_it_mun["RegionName"] = df_it_mun["RegionName"].replace(corrections) # Create a dictionary and replace incorrectly writen names
```

```python
def clean_column(name): # Column cleaning function
    # Replace any dashes, extra spaces, and unify apostrophes
    name = re.sub(r'[-\s]+', ' ', name)  # Replace dashes and multiple spaces with a single space
    name = re.sub(r"\'", "'", name)  # Normalize apostrophes (in case they are not consistent)
    name = name.strip()  # Remove leading/trailing spaces
    name = name.title()  # Convert to Title Case (optional, depending on your need)
    return name

# Apply the cleaning function to the MunicipalityName column and RegionName column
df_it_mun['MunicipalityName'] = df_it_mun['MunicipalityName'].apply(clean_column)
df_it_mun['RegionName'] = df_it_mun['RegionName'].apply(clean_column)

# Check the cleaned data
print(df_it_mun)
```

```python
df_it_mun["Pop2011"] = df_it_mun["Pop2011"].astype(str).str.extract(r"(\d+)") # Extract only numbers
# Identify non-numeric values in Pop2011, cast str type and check if there are only numerics (identifies non-numeric values)
non_numeric_values = df_it_mun[~df_it_mun['Pop2011'].astype(str).str.isnumeric()]
print(non_numeric_values)
```
