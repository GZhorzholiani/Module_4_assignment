```python
import pandas as pd
import re
```

```python
df_covid_region=pd.read_excel("../Datasets/Covid_region.xlsx", engine="openpyxl")
df_covid_region.head()
```

```python
df_covid_region.columns = df_covid_region.columns.str.strip()
df_covid_region.columns = df_covid_region.columns.str.replace(" ", "_").str.lower()
print(df_covid_region.columns)
```

```python
df_covid_region['data'] = df_covid_region['data'].str.split('T').str[0]
df_covid_region=df_covid_region.iloc[:,1:]
```

```python
df_covid_region["data"] = pd.to_datetime(df_covid_region["data"])
df_covid_region.head()
```
