```python
import pandas as pd
```

```python
# Read Covid Region file
df_covid_region=pd.read_excel("../Datasets/Covid_region.xlsx", engine="openpyxl")
df_covid_region.head()
```

```python
df_covid_region.columns = df_covid_region.columns.str.strip() # Remove leading/trailing spaces
df_covid_region.columns = df_covid_region.columns.str.replace(" ", "_").str.lower() # Replace spaces with underscore and lower every name
print(df_covid_region.columns)
```

```python
df_covid_region['data'] = df_covid_region['data'].str.split('T').str[0] # Split with 'T' and pick only first generated string
df_covid_region=df_covid_region.iloc[:,1:] # Remove the first column (index 0) from the dataframe by selecting all columns starting from index 1
df_covid_region.head()
```

```python
df_covid_region.isna().sum() # Check NA sum
```

```python
len(df_covid_region.data.unique()) # Check for unique
```

```python
len(df_covid_region.data.values) # Check for values
```

```python
print(df_covid_region['data'].value_counts().head(10)) # Check value counts
```

```python
df_covid_region_filtered = df_covid_region[df_covid_region['nome_regione'] == "Lazio"] # Filter by Lazio region
print(df_covid_region_filtered['data'].value_counts().head(10))
```

```python
df_covid_region.info()
```

```python
df_covid_region["data"] = pd.to_datetime(df_covid_region["data"]) # Convert data to datetime
df_covid_region.head()
```

```python
first_test_esen = df_covid_region[df_covid_region['test_eseguiti'] > 0] # Check where test_eseguiti is more than 0
first_test_esen.head()
```

```python
df_covid_region.info()
```

```python
df_covid_region['test_eseguiti'] = pd.to_numeric(df_covid_region['test_eseguiti'], errors='coerce').fillna(0).astype(int) # Converted to int and changed NA to 0
df_covid_region.columns = ['Date', 'Country', 'RegionCode', 'RegionName', 'Latitude', 'Longitude', 
                           'HospitalizedPatients', 'PatientsInIntensiveCare', 'TotalHospitalizedPatients', 
                           'HomeConfinement', 'CurrentPositiveCases', 'NewPositiveCases', 
                           'Healed', 'Dead', 'TotalPositiveCases', 'TestsExecuted'] # Change column names
df_covid_region.info()
```

```python
cleaned_df_covid_region = df_covid_region  # Finalized version
cleaned_df_covid_region.head()
```

```python
cleaned_df_covid_region.to_pickle("../Cleaned_datasets/cleaned_df_covid_region.pkl")
```
