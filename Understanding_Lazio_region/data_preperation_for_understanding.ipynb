```python
import pandas as pd
```

```python
df = pd.read_pickle("../Cleaned_datasets/cleaned_df_altitude_of_municipalities.pkl")
df_1 = pd.read_pickle("../Cleaned_datasets/cleaned_df_covid_province.pkl")
```

```python
# Filter for Region = Lazio i.e Regional Code = 12
Lazio_df = df[df["Region_Code"] == 12]
```

```python
# Count, in a new DataFrame, the number of municipalities (Name)
# and their average altitude (Average) in each province (Province_Code)

result = Lazio_df.groupby("Province_Code").agg(
    Municipality_Count=("Name", "count"),
    Average_Altitude=("Average", "mean")
).reset_index()
```

```python
print (result)
```

```python
df3 = df_1[df_1['Region']=='Lazio']
# Rename column in df3 (cleaned Lazio Covid_province) to match 'result' dataframe column name
df3 = df3.rename(columns={"ProvinceCode": "Province_Code"})
df3.head()
```

```python
# Merge Lazio Covid Province and result on province code to have the province name
merged = result.merge(df3, on="Province_Code")

merged.head()
```

```python
# Merge Lazio Covid Province and result on province code to have the province name
merged = result.merge(df3, on="Province_Code")

merged.head()
```

```python
merged_cov_prov_and_alt_mun = merged
merged_cov_prov_and_alt_mun.to_pickle("../Cleaned_datasets/merged_cov_prov_and_alt_mun.pkl")
```
