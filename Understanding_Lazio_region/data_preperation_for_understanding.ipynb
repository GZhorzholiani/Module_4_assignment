```python
import pandas as pd
```

```python
# Reading the files
cleaned_df_altitude_of_municipalities = pd.read_pickle("../Cleaned_datasets/cleaned_df_altitude_of_municipalities.pkl")
cleaned_df_covid_province = pd.read_pickle("../Cleaned_datasets/cleaned_df_covid_province.pkl")
```

```python
# Filter for Region = Lazio i.e Regional Code = 12
cleaned_df_altitude_of_municipalities_lazio = cleaned_df_altitude_of_municipalities[cleaned_df_altitude_of_municipalities["RegionCode"] == 12]
```

```python
# Count, in a new DataFrame, the number of municipalities (Name)
# and their average altitude (Average) in each province (Province_Code)
result = cleaned_df_altitude_of_municipalities_lazio.groupby("ProvinceCode").agg(
    MunicipalityCount=("MunicipalityName", "count"),
    AverageAltitude=("Average", "mean")
).reset_index()
print (result)
```

```python
# Filter for Region = Lazio
cleaned_df_covid_province_lazio = cleaned_df_covid_province[cleaned_df_covid_province['RegionName']=='Lazio']
cleaned_df_covid_province_lazio.head()
```

```python
# Merge Lazio Covid Province and result on province code to have the province name
merged = result.merge(cleaned_df_covid_province_lazio, on="ProvinceCode")
merged.head()
```

```python
# Merge Lazio Covid Province and result on province code to have the province name
merged = result.merge(cleaned_df_covid_province_lazio, on="ProvinceCode")
merged.head()
```

```python
merged_cov_prov_and_alt_mun = merged # Finalized Result
merged_cov_prov_and_alt_mun.to_pickle("../Cleaned_datasets/merged_cov_prov_and_alt_mun.pkl")
```
